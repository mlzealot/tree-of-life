```python
import numpy as np
import pandas as pd
```

```python
source_file_path, dest_file_path = "./src/data/Kidney_v1.1.csv", "./src/data/KidneyRegular.csv"
data = pd.read_csv(source_file_path)
```

```python
current_colnames = list(data.columns)

new_colnames = list(data.iloc[9, :])
data = data.iloc[10:, :]
data.rename(columns = {curr: new_names for curr, new_names in zip(current_colnames, new_colnames)}, 
            inplace=True)
data.reset_index(inplace=True)
data.drop("index", axis=1, inplace=True)
# data.replace(np.nan, None, inplace=True)
display(data.head())
```

```python
temp = list(data.loc[data["CT/1/ID"].isna() & (~(data["CT/1/LABEL"].isna())), "CT/1/LABEL"])
print(temp)
data.loc[data["CT/1/ID"].isna() & (~(data["CT/1/LABEL"].isna())), "CT/1/ID"] = temp
```

```python
def replaceNans(col_1, col_2):
    print("Col1, col2", str(col_1), str(col_2))
    if(str(col_1) == "nan" and (str(col_2) != "nan")):
        print(col_2)
    return str(col_2)
```

```python
data["CT/1/ID"] = data.apply(lambda x: replaceNans(x["CT/1/ID"], x["CT/1/LABEL"]), axis=1)
```
